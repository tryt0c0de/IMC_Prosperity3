```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prices_day__2 = pd.read_csv("/Users/difa/Desktop/IMC_Prosperity3/round-1-island-data-bottle/prices_round_1_day_-2.csv",sep = ";")
prices_day__2 = prices_day__2[prices_day__2["product"] == "KELP"]
prices_day__2["day"] = -2
prices_day__1 = pd.read_csv("/Users/difa/Desktop/IMC_Prosperity3/round-1-island-data-bottle/prices_round_1_day_-1.csv",sep = ";")
prices_day__1 = prices_day__1[prices_day__1["product"] == "KELP"]
prices_day__1["day"] = -1
prices_day_0 =pd.read_csv("/Users/difa/Desktop/IMC_Prosperity3/round-1-island-data-bottle/prices_round_1_day_0.csv",sep = ";")
prices_day_0 = prices_day_0[prices_day_0["product"] == "KELP"]
prices_day_0["day"] = 0
all_prices = pd.concat([prices_day__2, prices_day__1, prices_day_0])
```

```python
all_prices = all_prices.sort_values(by=['day', 'timestamp']).reset_index(drop=True)

# 2) Create a simple future return: next mid_price / current mid_price - 1
# Shift by -1 means we look one step ahead
all_prices['future_return'] = (all_prices['mid_price'].shift(-1) - all_prices['mid_price']) / all_prices['mid_price']

# Because of the shift, the last row won't have a future_return
all_prices.dropna(subset=['future_return'], inplace=True)

# 3) For classification, define an UP/DOWN label 
#    (1 if future_return > 0, 0 if future_return <= 0)
all_prices['future_return_label'] = (all_prices['future_return'] > 0).astype(int)
```

```python
all_prices['spread_1'] = all_prices['ask_price_1'] - all_prices['bid_price_1']

# Total bid volume (levels 1..3)
all_prices['total_bid_vol'] = (all_prices['bid_volume_1'].fillna(0) +
                               all_prices['bid_volume_2'].fillna(0) +
                               all_prices['bid_volume_3'].fillna(0))

# Total ask volume (levels 1..3)
all_prices['total_ask_vol'] = (all_prices['ask_volume_1'].fillna(0) +
                               all_prices['ask_volume_2'].fillna(0) +
                               all_prices['ask_volume_3'].fillna(0))

# Order book volume imbalance
all_prices['volume_imbalance'] = (
    all_prices['total_bid_vol'] / (all_prices['total_bid_vol'] + all_prices['total_ask_vol'] + 1e-9)
)

# Rolling feature: for example, a 5-row rolling average of mid_price
all_prices['mid_price_rolling_mean'] = all_prices['mid_price'].rolling(window=5).mean()
all_prices['mid_price_rolling_mean'] = all_prices['mid_price_rolling_mean'].bfill()

# Momentum-like feature
all_prices['momentum'] = all_prices['mid_price'] - all_prices['mid_price_rolling_mean']
```

```python
feature_cols = [
    'spread_1',
    'total_bid_vol',
    'total_ask_vol',
    'volume_imbalance',
    'momentum'
    # ... add more if you like
]

# Prepare X and y for both regression and classification
X = all_prices[feature_cols].copy().dropna()

# For a REGRESSION target:
y_reg = all_prices['future_return'].copy().dropna()

# For a CLASSIFICATION target:
y_clf = all_prices['future_return_label'].copy().dropna()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error, r2_score

# === LOAD DATA ===
dfs = []
for day in [-2, -1, 0]:
    df = pd.read_csv(f"/Users/difa/Desktop/IMC_Prosperity3/round-1-island-data-bottle/prices_round_1_day_{day}.csv", sep=";")
    df = df[df["product"] == "KELP"]
    df["day"] = day
    dfs.append(df)

df = pd.concat(dfs).sort_values(["day", "timestamp"]).reset_index(drop=True)

# === FEATURE ENGINEERING ===
df["mid_price"] = (df["ask_price_1"] + df["bid_price_1"]) / 2
df["spread_1"] = df["ask_price_1"] - df["bid_price_1"]
df["total_bid_vol"] = df[["bid_volume_1", "bid_volume_2", "bid_volume_3"]].fillna(0).sum(axis=1)
df["total_ask_vol"] = df[["ask_volume_1", "ask_volume_2", "ask_volume_3"]].fillna(0).sum(axis=1)
df["volume_imbalance"] = df["total_bid_vol"] / (df["total_bid_vol"] + df["total_ask_vol"] + 1e-9)
df["mid_price_rolling_mean"] = df["mid_price"].rolling(5).mean().bfill()
df["momentum"] = df["mid_price"] - df["mid_price_rolling_mean"]
df["future_return_5"] = (df["mid_price"].shift(-5) - df["mid_price"]) / df["mid_price"]

# === DROP NAs ===
feature_cols = [
    "spread_1", "total_bid_vol", "total_ask_vol", 
    "volume_imbalance", "momentum"
]
df = df.dropna(subset=feature_cols + ["future_return_5"])

X = df[feature_cols].copy()
y = df["future_return_5"]

# === SPLIT DATA ===
train_size = int(0.8 * len(df))
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]
mid_price_test = df.iloc[train_size:]["mid_price"].values
mid_price_future = df.iloc[train_size+5:]["mid_price"].values[:len(y_test)]

# === TRAIN MODEL ===
model = RidgeCV(alphas=[0.01, 0.1, 1.0, 10.0])
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# === BACKTEST STRATEGY ===
pnl = []
for pred, now, later in zip(y_pred, mid_price_test, mid_price_future):
    if pred > 0.001:
        pnl.append(later - now)
    elif pred < -0.001:
        pnl.append(now - later)
    else:
        pnl.append(0)

cumulative_pnl = np.cumsum(pnl)
sharpe = np.mean(pnl) / (np.std(pnl) + 1e-9)

# === REPORT ===
print("\n==== MODEL REPORT ====")
print("R^2 Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("Sharpe Ratio:", sharpe)
print("Cumulative PnL:", cumulative_pnl[-1])

print("\n==== COEFFICIENTS ====")
for name, coef in zip(feature_cols, model.coef_):
    print(f"{name}: {coef}")
print("Intercept:", model.intercept_)

# === PLOT PNL ===
plt.plot(cumulative_pnl)
plt.title("Cumulative PnL (Simple Strategy)")
plt.xlabel("Trade #")
plt.ylabel("PnL")
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LassoCV, ElasticNetCV
from sklearn.metrics import mean_squared_error, r2_score

# === Load data ===
dfs = []
for day in [-2, -1, 0]:
    df = pd.read_csv(f"/Users/difa/Desktop/IMC_Prosperity3/round-1-island-data-bottle/prices_round_1_day_{day}.csv", sep=";")
    df = df[df["product"] == "KELP"]
    df["day"] = day
    dfs.append(df)

df = pd.concat(dfs).sort_values(["day", "timestamp"]).reset_index(drop=True)

# === Core feature engineering ===
df["mid_price"] = (df["ask_price_1"] + df["bid_price_1"]) / 2
df["spread_1"] = df["ask_price_1"] - df["bid_price_1"]
df["total_bid_vol"] = df[["bid_volume_1", "bid_volume_2", "bid_volume_3"]].fillna(0).sum(axis=1)
df["total_ask_vol"] = df[["ask_volume_1", "ask_volume_2", "ask_volume_3"]].fillna(0).sum(axis=1)
df["volume_imbalance"] = df["total_bid_vol"] / (df["total_bid_vol"] + df["total_ask_vol"] + 1e-9)
df["mid_price_rolling_mean"] = df["mid_price"].rolling(5).mean().bfill()
df["momentum"] = df["mid_price"] - df["mid_price_rolling_mean"]
df["future_return_5"] = (df["mid_price"].shift(-5) - df["mid_price"]) / df["mid_price"]

# === Extra features ===
df["spread_change"] = df["spread_1"].diff()
df["momentum_lag1"] = df["momentum"].shift(1)
df["price_change"] = df["mid_price"].diff()
df["volatility_10"] = df["mid_price"].rolling(10).std().bfill()
df["bid_ask_ratio"] = df["total_bid_vol"] / (df["total_ask_vol"] + 1e-9)

# === Define all features ===
feature_cols = [
    "spread_1", "total_bid_vol", "total_ask_vol", "volume_imbalance", "momentum",
    "spread_change", "momentum_lag1", "price_change", "volatility_10", "bid_ask_ratio"
]

# === Clean and split ===
df = df.dropna(subset=feature_cols + ["future_return_5"])
X = df[feature_cols]
y = df["future_return_5"]

train_size = int(0.8 * len(df))
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]
mid_price_test = df.iloc[train_size:]["mid_price"].values
mid_price_future = df.iloc[train_size+5:]["mid_price"].values[:len(y_test)]

# === Train models ===
print("Training LassoCV...")
lasso = LassoCV(cv=5).fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)

print("Training ElasticNetCV...")
elastic = ElasticNetCV(cv=5).fit(X_train, y_train)
y_pred_elastic = elastic.predict(X_test)

# === Evaluation + Backtest ===
def evaluate(name, y_pred):
    pnl = []
    for pred, now, later in zip(y_pred, mid_price_test, mid_price_future):
        if pred > 0.001:
            pnl.append(later - now)
        elif pred < -0.001:
            pnl.append(now - later)
        else:
            pnl.append(0)
    pnl = np.array(pnl)
    cum_pnl = np.cumsum(pnl)
    sharpe = np.mean(pnl) / (np.std(pnl) + 1e-9)
    
    print(f"\n== {name} ==")
    print("R² Score:", r2_score(y_test, y_pred))
    print("MSE:", mean_squared_error(y_test, y_pred))
    print("Sharpe:", sharpe)
    print("Cumulative PnL:", cum_pnl[-1])
    
    plt.plot(cum_pnl)
    plt.title(f"{name} - Cumulative PnL")
    plt.xlabel("Trade")
    plt.ylabel("PnL")
    plt.grid(True)
    plt.show()
    
    return pnl, cum_pnl

# === Coefficients ===
def print_model_info(name, model):
    print(f"\n{name} Coefficients:")
    for col, coef in zip(feature_cols, model.coef_):
        print(f"{col}: {coef:.8f}")
    print("Intercept:", model.intercept_)

# === Run all ===
evaluate("Lasso", y_pred_lasso)
print_model_info("Lasso", lasso)

evaluate("ElasticNet", y_pred_elastic)
print_model_info("ElasticNet", elastic)

```
