```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Merge data from files for days -2, -1, and 0
days = [-2, -1, 0]
data_frames = []
for day in days:
    file_path = f'/Users/andaks/Columbia/Side projects/IMC_Prosperity3/Round1/round-1-island-data-bottle/prices_round_1_day_{day}.csv'
    temp_data = pd.read_csv(file_path, delimiter=';')
    data_frames.append(temp_data)

# Concatenate all data into a single DataFrame
data = pd.concat(data_frames, ignore_index=True)
# Filter the data for the product 'KELP'
data_kelp = data[data['product'] == 'KELP']
# Calculate midprice for each tick
data_kelp['mid_price'] = (data_kelp['bid_price_1'] + data_kelp['ask_price_1']) / 2

# Calculate tick returns
data_kelp['tick_return'] = data_kelp['mid_price'].pct_change()

# Plot the distribution of tick returns
plt.hist(data_kelp['tick_return'].dropna(), bins=100, edgecolor='k', alpha=1)
plt.title('Distribution of Tick Returns')
plt.xlabel('Tick Return')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt

# Merge data from files for days -2, -1, and 0
days = [-2, -1, 0]
data_frames = []
for day in days:
    file_path = f'/Users/andaks/Columbia/Side projects/IMC_Prosperity3/Round1/round-1-island-data-bottle/prices_round_1_day_{day}.csv'
    temp_data = pd.read_csv(file_path, delimiter=';')
    data_frames.append(temp_data)

# Concatenate all data into a single DataFrame
data = pd.concat(data_frames, ignore_index=True)
# Filter the data for the product 'KELP'
data_squid = data[data['product'] == 'SQUID_INK']
# Calculate midprice for each tick
data_squid['mid_price'] = (data_squid['bid_price_1'] + data_squid['ask_price_1']) / 2

# Calculate tick returns
data_squid['tick_return'] = data_squid['mid_price'].pct_change()

# Plot the distribution of tick returns
plt.hist(data_squid['tick_return'].dropna(), bins=100, edgecolor='k', alpha=1)
plt.title('Distribution of Tick Returns')
plt.xlabel('Tick Return')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data_squid.index, data_squid['mid_price'], label='Mid Price SQUID INK', color='blue')
plt.plot(data_kelp.index, data_kelp['mid_price'], label='Mid Price KELP', color='red')
plt.title('Dynamic of Mid Price Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python

merged_returns = pd.merge(
    data_squid[['day', 'timestamp', 'tick_return']], 
    data_kelp[['day', 'timestamp', 'tick_return']], 
    on=['day', 'timestamp'], 
    suffixes=('_squid', '_kelp')
)
merged_returns = merged_returns.dropna()
```

```python
merged_returns.head(10)
```

```python
# Calculate rolling correlation
rolling_corr = merged_returns['tick_return_squid'].rolling(window=1000).corr(merged_returns['tick_return_kelp'])

# Plot rolling correlation
plt.figure(figsize=(12, 6))
plt.plot(rolling_corr, label='Rolling Correlation (window=10000)', color='purple')
plt.title('Rolling Correlation Between Squid Ink and Kelp Tick Returns')
plt.xlabel('Index')
plt.ylabel('Correlation')
plt.legend()
plt.grid(True)
plt.show()

# Calculate and print overall correlation
overall_correlation = merged_returns['tick_return_squid'].corr(merged_returns['tick_return_kelp'])
print(f"Overall Correlation: {overall_correlation}")
```
