```python
import pandas as pd

import matplotlib.pyplot as plt

# Merge data from files for days -2, -1, and 0
days = [-2, -1, 0]
data_frames = []
for day in days:
    file_path = f'/Users/andaks/Columbia/Side projects/IMC_Prosperity3/Round1/round-1-island-data-bottle/prices_round_1_day_{day}.csv'
    temp_data = pd.read_csv(file_path, delimiter=';')
    data_frames.append(temp_data)

# Concatenate all data into a single DataFrame
data = pd.concat(data_frames, ignore_index=True)
# Filter the data for the product 'KELP'
data_squid = data[data['product'] == 'SQUID_INK']
# Calculate midprice for each tick
data_squid['mid_price'] = (data_squid['bid_price_1'] + data_squid['ask_price_1']) / 2

# Calculate tick returns
data_squid['tick_return'] = data_squid['mid_price'].pct_change()

# Plot the distribution of tick returns
plt.hist(data_squid['tick_return'].dropna(), bins=100, edgecolor='k', alpha=1)
plt.title('Distribution of Tick Returns')
plt.xlabel('Tick Return')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data_squid.index, data_squid['mid_price'], label='Mid Price SQUID INK', color='blue')
plt.plot(data_kelp.index, data_kelp['mid_price'], label='Mid Price KELP', color='red')
plt.title('Dynamic of Mid Price Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.legend()
plt.grid(True)
plt.show()
# Calculate total trading volume for each timestamp
data_squid['total_volume'] = data_squid[['bid_volume_1', 'bid_volume_2', 'bid_volume_3']].sum(axis=1) / data_squid[['ask_volume_1', 'ask_volume_2', 'ask_volume_3']].sum(axis=1)
data_kelp['total_volume'] = data_kelp[['bid_volume_1', 'bid_volume_2', 'bid_volume_3']].sum(axis=1) / data_kelp[['ask_volume_1', 'ask_volume_2', 'ask_volume_3']].sum(axis=1)

# Plot total trading volume
plt.figure(figsize=(12, 6))
plt.plot(data_squid.index, data_squid['total_volume'], label='Total Volume SQUID INK', color='green', alpha=0.6)
plt.plot(data_kelp.index, data_kelp['total_volume'], label='Total Volume KELP', color='orange', alpha=0.6)
plt.title('Dynamic of Total Trading Volume Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Total Trading Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF for squid tick returns
plt.figure(figsize=(12, 6))
plot_acf(data_squid['tick_return'].dropna(), lags=50, alpha=0.05, title="ACF of Squid Tick Returns")
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(data_squid['tick_return'].dropna(), lags=50, alpha=0.05, title="PACF of Squid Tick Returns")
plt.show()
```

```python
# Calculate the moving average for the last 1000 periods
data_kelp['moving_avg'] = data_kelp['mid_price'].rolling(window=5000).mean()

# Plot mid price and moving average
plt.figure(figsize=(12, 6))
plt.plot(data_kelp.index, data_kelp['mid_price'], label='Mid Price', color='blue', alpha=0.6)
plt.plot(data_kelp.index, data_kelp['moving_avg'], label='Moving Average (1000 periods)', color='orange', alpha=0.8)
plt.title('Mid Price and Moving Average for KELP')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Calculate log returns
log_returns = np.log(data_kelp['mid_price']).diff()

# Calculate drift (r) and variance coefficient (sigma^2)
r = log_returns.mean()
sigma_squared = log_returns.var()

# Define time (T) as the range of the data
T = np.arange(len(data_kelp))

# Calculate the line values
S_0 = 2000  # Initial price
line_values = S_0 * np.exp(r * T + 0.5 * sigma_squared * T)

# Plot the price dynamics and the line
plt.figure(figsize=(12, 6))
plt.plot(data_kelp.index, data_kelp['mid_price'], label='Mid Price', color='blue', alpha=0.6)
plt.plot(data_kelp.index, line_values, label='S_0 * exp(r * T + 1/2 * sigma^2 * T)', color='orange', linestyle='--')
plt.title('Mid Price Dynamics with Drift and Variance Line')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
stock_count = 0
# Create a dynamic plot for PnL
pnl_values = []
stock_pos = []

for i in range(len(data_kelp)):
    if data_kelp['mid_price'].iloc[i] <= line_values[i]:
        if stock_count < 50:  # Limit long position to 50
            pnl += data_kelp['mid_price'].iloc[i]
            stock_count += 1
    elif data_kelp['mid_price'].iloc[i] >= 1.0005 * line_values[i]:
        if stock_count > -50:  # Limit short position to -50
            pnl -= data_kelp['mid_price'].iloc[i]
            stock_count -= 1
    pnl_values.append(pnl)
    stock_pos.append(stock_count)

plt.figure(figsize=(12, 6))
plt.plot(data_kelp.index, pnl_values, label='PnL Over Time', color='green')
plt.title('Dynamic PnL Over Time')
plt.xlabel('Timestamp')
plt.ylabel('PnL')
plt.legend()
plt.grid(True)
plt.show()

print(f"Total PnL: {pnl}")

plt.figure(figsize=(12, 6))
plt.plot(data_kelp.index, stock_pos, label='PnL Over Time', color='green')
plt.title('Dynamic PnL Over Time')
plt.xlabel('Timestamp')
plt.ylabel('PnL')
plt.legend()
plt.grid(True)
plt.show()

print(f"Total PnL: {pnl}")
```
